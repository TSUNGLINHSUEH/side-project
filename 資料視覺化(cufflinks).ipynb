透過cufflinks，可以結合pandas與plot，並讓圖表可以透過游標移動顯示資料數值

```python
import pandas as pd
import numpy as np
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
print(__version__)
import cufflinks 
import sys

init_notebook_mode(connected = True)
cufflinks.go_offline()

df = pd.DataFrame(np.random.randn(100, 4), columns = "A B C D".split())
df
```

```python
df2 = pd.DataFrame({"Category":["Z", "x", "y"], "Values":[5, 6, 7]})
df2
```

■使用plotly及cufflinks

散佈圖(scatter)

```python
#iplot(kind = "scatter", x, y, mode = "markers")。mode沒有設定會變成所有點會連成一條直線。
df.iplot(kind = "scatter", x = "A", y = "B", mode = "markers", size = 20)
```

直條圖(barplots)

```python
#iplot(kind = "bar", x, y)
df2.iplot(kind = "bar" , x = "Category" , y = "Values")

#iplot前也可以用groupby或pandas常用的數學運算
df.sum().iplot(kind = "bar")
```

盒鬚圖(boxplots)

```python
df.iplot(kind = "box")
```

3D曲線圖(3D surface)

```python
#iplot(kind = "surface", colorscale = "")colorscale改變Z軸顏色
df3 = pd.DataFrame({"x":[5, 6, 7, 8, 9], "y":[50, 60, 70, 80, 90],"z":[500, 600, 700, 800, 900]})
df3.iplot(kind = "surface", colorscale = "rdylbu")
```

直方圖(histgram)

```python
df.iplot(kind = "hist")
```

spread plot

```python
#下面的區域是前者減掉後者的數值，大於0會出現在上面，小於0會出現在下面，可以看出前者大於後者的次數多還是少
df[["A", "B"]].iplot(kind = "spread")
```

泡泡圖

```python
df.iplot(kind = "bubble", x = "A", y = "B", size = "C")
```
